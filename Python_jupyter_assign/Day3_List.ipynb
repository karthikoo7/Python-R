# LIST COMPREHENSION -- VERY IMP

```python
lst = [x for x in range(0,101)]
print(lst)
```

```python
l1 = [1,2,3,4,5,6]
l2 = [x**2 for x in l1]
print(l2)
```

```python
##lst = [x%7==0 for x in range(0,101)]
lst = [x for x in range(0,101) if x%7 ==0]
print(lst)
```

```python
l1 = ['red','blue']
l2 = ['house','car']

l3 = [(x,y) for x in l1 for y in l2]
print(l3)
```

### Assignments:

```python
a = [1,3,4,5,6,7,27]
sum = 0
length = len(a)
for i in a:
    sum += i

print("AVG",sum/length)
```

```python
a = [1,3,4,5,6,7,27,-24,-36,-1,22,44,66]
sum = 0
for i in a:
    sum += i
print(sum)
```

```python
a = [1,3,4,5,66,6,7,27]
largestEven = 0;
largestOdd = 0;
for i in a:
    if i%2 == 0:
        if i>largestEven:
            largestEven = i
    elif i%2 != 0:
        if i>largestOdd:
            largestOdd = i
print("EVEN: ", largestEven, " Odd: ",largestOdd)
```

```python
name = 'Raju'
age= 22
marks = 92.3

# using percent
print("Hello my name is %s, and I am %d years old. I scored %f marks in exams"%(name,age,marks))
#using format method -default formatting
print("Hello my name is {}, and I am {} years old. I scored {} marks in exams".format(name,age,marks))
#using format method - Positional formatting 
print("Hello my name is {0},And {0} is the greatest,and I am {1} years old. I scored {2} marks in exams".format(name,age,marks))
#using format method - Keyword formatting
print("Hello my name is {n}, and I am {a} years old. I scored {m} marks in exams".format(n=name,a=age,m=marks))
# using format string
print(f"Hello my name is {name}, and I am {age} years old. I scored {marks} marks in exams")

```

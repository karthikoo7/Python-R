```python
#reversed string 

name = input('Naam: ')
count = 0
for i in name:
    count += 1

for i in range(count-1,-1,-1):
    print(name[i],end='')
```

```python
n = int(input('Enter'))
facto = 1
for i in range(1,n+1):
    facto *= i;
print(facto)
```

```python
a = int(input('Num: '))
charge = 0
bill = 0
i=0
while i<=a:
    if i == 101:
        charge = 5
    elif i == 201:
        charge = 10

    bill += charge
    i += 1

print("Units: ",a," Bill: ",bill)
```

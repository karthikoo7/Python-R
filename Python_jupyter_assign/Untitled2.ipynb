## Panda Assignments


```python
import pandas as pd

```

```python
l1 = [1,2,3,4,5]
l2 = [10,20,30,40,50]

s1 = pd.Series(l1)
s2 = pd.Series(l2)

print(s1-s2)
print('*'*10)

print(s1+s2)
print('*'*10)

print(s1*s2)
print('*'*10)

print(s1/s2)
print('*'*10)
```

```python
d = {'a':10,'b':20,'c':30,'d':40}
s1 = pd.Series(d)
print(s1)
```

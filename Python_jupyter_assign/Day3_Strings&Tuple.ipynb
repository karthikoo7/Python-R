```python
a = 'AbCdRdRG'
countUpper = 0
countLower = 0
for i in a:
    if i.isupper():
        countUpper +=1
    elif i.islower():
        countLower +=1

else:
    print(countUpper,countLower)
```

```python
vow = 'aeiou'
count = 0
name = 'Karthik'
for i in name:
    if i not in vow:
        count += 1

print(count)
```

```python
a = 'Hello I am 1st but my friends are 2nd and 3rd'

number = 0

for i in a:
    if i.isdigit():
        number += int(i)

print(number)
```

```python
a = 'abc123 cds'

number = 0

for i in a:
    if i.isalpha():
        number += 1

print(number)
```

## NUmpy assignments


```python
import numpy as np
a = np.arange(5)
print(a)

## number of elements in the array == product of the shape of array i.e rows*col for 2D arrays | 3D rows*cols*no of arrays
# 2d Arrays
a1 = np.arange(1,21).reshape(4,5)
print(a1)

# 3d arrays
a2 = np.arange(1,21).reshape(1,4,5)
print(a2)

## Attributes

print('*'*40)
print(a.ndim)
print(a.shape)
print(a.size)
print(a.dtype)

print('*'*40)
print(a1.ndim)
print(a1.shape)
print(a1.size)
print(a1.dtype)

print('*'*40)
print(a2.ndim)
print(a2.shape)
print(a2.size)
print(a2.dtype)

```

```python
# using empty() -- has garbage values , only allocates memeory
a = np.empty((3,5))
print(a)
print(a.shape)

## create array and fill with specified values
print('*'*40)
a1 = np.full((3,5),10)
print(a1)

# creates and assigns float value 0.
print('*'*40)
a2 = np.zeros((3,5))
print(a2)

# creates and assigns float value 1.
print('*'*40)
a2 = np.ones((3,5))
print(a2)

```

```python
# using arange to create array with value in a range
a = np.arange(14,54)
print(a)
```

```python
# creating using arange and then using reshape to add dimensions
a = np.arange(1,41)
print(a)
print('*'*40)
a1 = a.reshape((4,10))
print(a1)
```

```python
# operations on 2D array
a = np.arange(1,16).reshape(5,3)
print(a)

print('*'*40)
print(a+4)

print('*'*40)
print(a*2)

print('*'*40)
print(a-3)

print('*'*40)
print(a/2)
```

```python
# OPerations between two 2D Arrays
a = np.arange(1,10).reshape(3,3)
print(a)

a2 = np.arange(11,20).reshape(3,3)
print(a2)

print('*'*40)
print(a+a2)

print('*'*40)
print(a-a2)

print('*'*40)
print(a.dot(a2))

print('*'*40)
print(a/a2)

# using sum(),min(),max() on 2D array and axis=0/axis=1
print('*'*40)
print(a)
print('-'*40)
print(a.sum(axis=0))
print(a.max(axis=0))
print(a.min(axis=0))


# using transpose
print('*'*20)
a2.transpose()
```

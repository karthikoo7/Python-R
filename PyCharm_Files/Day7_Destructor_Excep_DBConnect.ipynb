def decorator(func):
    def wrapper(*args,**kwargs):
        print('Before')
        func(*args,**kwargs)
        print('After')
    return wrapper





```python
def decorator(func):
    def wrapper(no):
        print('Before')
        res =func(no)
        print("after")
        return res
    return wrapper

@decorator
def square(n):
    return n**2

@decorator
def cube(n):
    return n**3

print(cube(3))
print(square(2))
```

## Decorator Assignments:

```python
def simpleDecorator(func):
    def wrapper():
        print("Before Function")
        func()
        print("After Function")
    return wrapper

@simpleDecorator
def func():
    pass

func()
```

```python
import time

def calculateTime(func):
    def wrapper():
        start = time.time()
        func()
        print("After")
        end = time.time()
        print("Time taken: ",end - start)
    return wrapper

@calculateTime
def slow_function():
    for i in range(10):
        print(i**2)

slow_function()
```

```python
def to_uppercase(func):
    def wrapper():
        res2 = ''
        res = func()
        res2 = res.upper()
        print("Before: ",res)
        return res2
    return wrapper


@to_uppercase
def getMessage():
    return "Hello there Gentlemen!"

print(getMessage())
```

```python
def checkAdmin(func):
    def wrapper(role):
        if role != 'Admin':
            print("Only Admin Allowed")
        else:
            func(role)
    return wrapper

@checkAdmin
def deletedata(role):
    print("Data Deleted")

deletedata("Admin")
```

```python
class typerError(Exception):
    def __init__(self,msg):
        self.msg = msg

    def __str__(self):
        return f"{self.msg}"

while True:
    try:
        Age = int(input("Enter Age"))
        Height = float(input("Height"))
        Weight = float(input("Weight"))
        name = input('Name: ')
        surname = input("Surname : ")

        if type(Age) != int:
            raise typerError("Age should be int")
        if type(Height) != float and type(Weight) != float:
            raise typerError("Values of height and weight shuld be float")
        if type(name) != str and type(surname) != str:
            raise typerError("Values of name and suranme should be string")

    except Exception as e:
        print(e)
    else:
        print("Good Job Soldier")
        break

```

```python
##Return odd values from matrix -- method 1
a1 = np.where(a%2 != 0)
print(a[a1])

# Return only even values from matrix -- method 2
print(a[a%2 ==0])
```

```python
# replace all even values with -1 in a matrix

print(a)
print('*'*40)
res = np.where(a%2==0,-1,a)
print(res)
```

```python
## from an array of random integers extract prime numbers
import math
a3 = np.random.randint(1,17,16)
print(a3)

## practice isPrime logic
def isPrime(n):
    if n>1:
        for i in range(2,int(math.sqrt(n)+1)):
            if n%i ==0:
                 return False

        return True
    else:
        return False


l2 = list(filter(isPrime,a3))
print(l2)
```

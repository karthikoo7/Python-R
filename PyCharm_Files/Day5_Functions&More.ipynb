## Types of Arguments:

```python
#variable length: *args and **kwargs
def add(*b):
    sum =0
    for i in b:
        sum+=i
    return sum

add(5,4,6,7,9,0)

## **kwargs -- keyword arguments

def subjects(**sub):
    for s,v in sub.items():
        print(s,v)


subjects(id = 12,name = 'Ravi',classed=12,paygrade=123123)
```

## Assignments: MAP AND FILTER FUNCTION

```python
def factorial(a = 1):
    fact = 1
    for i in range(1,a+1):
        fact *= i

    return fact

print(factorial(5))
```

```python
## IMP LOGIC

def checPrime(a):
    ans = 'Prime'
    for i in range(2,int(a**0.5)+1): ## range from 2 to square root of a
        if a%i == 0:
            ans = 'Not Prime'
    print(ans)

checPrime(2)
checPrime(4)
checPrime(7)
checPrime(9)
```

```python
sentence = 'Hi my name is Karthikeyan sharma'

def countWord(sent):
    count = 0
    for i in sent:
        if str.isalpha(i):
            count += 1
    return count;

print(countWord(sentence))
```

## ADVANCED QUERY ASSIGNMENT FOR MAP AND FILTER FN


```python
students = [
    {"name": "Amit", "marks": 78, "age": 20},
    {"name": "Sneha", "marks": 92, "age": 22},
    {"name": "Rahul", "marks": 65, "age": 19},
    {"name": "Priya", "marks": 88, "age": 21},
    {"name": "John", "marks": 55, "age": 20}
]
employees = [
    {"name": "Raj", "salary": 25000},
    {"name": "Simran", "salary": 40000},
    {"name": "Aman", "salary": 18000},
    {"name": "Neha", "salary": 32000}
]

names = list(map(lambda x:x['name'],students))
print('names:', names)

great = list(filter(lambda x:x['marks']>75,students))
print('Marks > 75',great)

scholar = list(filter(lambda x : x['marks']>=85,students))
print('Scholars: ',scholar)

fail = list(filter(lambda x : x['marks']<60,students))
print('Failed ',fail)

## Writing two conditions in lambda Expression
specialName = list(filter(lambda x:x['age']>20 and x['marks']>80,students))
print('SPECIAL : ',specialName)


salary = list(filter(lambda x: x['salary']>30000,employees))
print('HIGH Sal: ',salary)

lowsal = list(filter(lambda x:x['salary']<25000,employees))
print('Low SAl' , lowsal)

## Special QS Observe usage of .startswith() method on the value returned by x['name']
specialFilter = list(filter(lambda x:x['salary']>20000 and x['name'].startswith('N'),employees))
print('Name starts with N ',specialFilter)
```

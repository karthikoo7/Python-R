```python
class Rectangle:
    def __init__(self,length=1,breadth=1):
        self.rlen = length
        self.rbre = breadth

    def area(self):
        return self.rlen * self.rbre

    def perimeter(self):
        return 2*(self.rlen+self.rbre)

    def display(self):
        print('Area: ',self.area(),' ,Perimeter:',self.perimeter())

r = Rectangle()
r1 = Rectangle(10,25)

r.display()
r1.display()
```

```python
class bankAccount:
    def __init__(self,accountNo=1,name='Customer',balance=1):
        self.accNo = accountNo
        self.aname = name
        self.acBalance = balance

    def deposit(self,amount):
        self.acBalance += amount
        print('Deposit Success, Your Balance: ',self.acBalance)

    def withdraw(self,amount):
        if amount > self.acBalance:
            print('Cant Withdraw Insufficient Balance')
        else:
            self.acBalance -= amount
            print('WithDraw Success, Your Balance: ',self.acBalance)

    def display(self):
        print('Name:',self.aname,',AccNo : ',self.accNo,',Your Balance: ',self.acBalance)


ba = bankAccount(1234,'karthikeyan',10000)

ba.display()
ba.withdraw(5000)
ba.display()
ba.deposit(2000)
ba.withdraw(7500)
ba.display()
```

```python
class Person:
    def __init__(self,id=1,name=" k",email="sdfs",mobile=0000):
        self.pid = id
        self.pname = name
        self.emailId = email
        self.mobileNo = mobile

    def display(self):
        print(f'Person Details: {self.pid}, {self.pname}, {self.emailId},{self.mobileNo} ')


p = Person(1,'Karthikeyan','karthikeyansharma02636@gmail.com',9175506928)
p.display()

print('*'*10)
class Emp(Person):
    def __init__(self,dept=00,desg='dsf',sal=0000,id=0,name='k',email='sdfsd',mobile=0000):
        super().__init__(id,name,email,mobile)
        self.dept = dept
        self.desg = desg
        self.sal = sal

    def calculateNetSal(self):
        Netsal = self.sal+(self.sal*0.10)+(self.sal*0.15)-(self.sal*0.05)
        return f'Netsal: {Netsal}'

    def display(self):
        super().display()
        print(f'Salary Total: {self.sal},{self.desg},{self.dept} , Total : {self.calculateNetSal()}',)


e = Emp(10,'Manager',100,12,'Jonathan','Karthk@123',9175506928)
e.display()

print('*'*10)

class Member(Person):
    def __init__(self,membertype='sdfds',amtpaid=324234,id=1,name=" k",email="sdfs",mobile=0000):
        super().__init__(id,name,email,mobile)
        self.membertype = membertype
        self.amtpaid = amtpaid

    def display(self):
        super().display()
        print(f'Members: {self.membertype},{self.amtpaid}',)

m = Member('Good',123123,12,'Jonathan','Karthk@123',9175506928)

m.display()

print('*'*10)



```

```python
class Shape:
    def calculateArea(self,*b):
        pass

class circle(Shape):
    def calculateArea(self,r=1):
        return 3.14 * r*r


class Square(Shape):
    def calculateArea(self,a=1):
        return 4 * a

class Rectangle(Shape):
    def calculateArea(self,l=1,b=1):
        return l*b

c = circle()
print(c.calculateArea(7))

r = Rectangle()
print(r.calculateArea(2,4))

s = Square()
print(s.calculateArea(4))

## MRO OBject
print(Square.mro())
```

## Dictionary Comprehension

```python
d1 = {x:x**2 for x in range(1,15)}
print(d1)

print('*' * 100)

d2  = {x:(x**2,x**3) for x in range(0,20) if x%2 == 0}
print(d2)
```

```python

d1 = {}
n = int(input('Enter number of keys you want: '))
for i in range(0,n):
    key = int(input('Enter key: '))
    value = int(input('Enter Value: '))
    d1[key] = value

print(d1)
```

```python
d1 = {1:11,2:22,3:33,4:44}
d2 = {10:2,20:4,30:5,60:3,44:23}

d3 = {**d1,**d2}
print(d3)
```

```python
d3 = {x:(x,x**3) for x in range(1,5)}
print(d3)
```

```python
d1 = {'a':1,'b':2,'c':4,'d':6}
mul = 1
for i in d1.values():
    mul *= i
print(mul)
```

```python
d1 = {'a':1,'b':2,'c':4,'d':6}

delkey = input('Enter Key to delete: ')
new_key = input('Enter new Key')
if delkey in d1:
    d1[new_key] = d1.pop(delkey) ## new key with old value appende at the end
else:
    print('Key does not exist')


print(d1)
```

```python
## Menu Driven
prompt = input('Enter start to start :')
d1 = {'m':'math','p':'python','aj':'advanced java'}
while prompt.lower() == 'start':
    option = input('Commands : add, update,delete(key)')
    if option.lower() == 'add':
        key = input('enter key ')
        value = input('enter value ')
        d1[key] = value
    elif option.lower() == 'update':
        key = input('enter key ')
        value = input('enter value ')
        d1[key] = value
    elif option.lower() == 'delete':
        key = input('enter key ')
        print('Deleted:' ,d1.pop(key))
    elif option == 'quit':
        break

print('Thank you')
print(d1)
```

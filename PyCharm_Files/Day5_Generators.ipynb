## Generators **************** Generators VS Functions

```python
## Generator basics
def getValue():
    print('A')
    yield 10
    print('B')
    yield 20
    print('C')
    yield 30

g = getValue()
g2 = getValue()
print(g)

print(type(g))

print(next(g))
print(next(g))
print(next(g))

##OR

for i in g2:
    print(i)
```

```python
def sqrGen():
    for i in range(1,101):
        yield i**2

g = sqrGen() ## storing generator object in g

for i in range(1,11):
    print(next(g)) ## execution till square of 10

print('***********************************')
print('HELLO WORLD')
print('EXECUTION PAUSED')
print('GENERATOR DEMO')
print('***********************************')


for i in range(1,11):
    print(next(g))## execution continued from square of 11 to square of 20
```

```python
def even_number(n):
    for i in range(0,n):
        if i%2 == 0:
            yield i


g = even_number(10)


print(next(g)) # execution till 0 [method 1 of doing the same thing]
print('*'*5)
for i in g : # state maintained is 0 and continued from thereafter [method 2 of doing the same thing]
    print(i)
```

```python
def generateSquares(n):
    for i in range(1,n+1):
        yield i*i

g = generateSquares(10)
 #[method 1 of doing the same thing]
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print('*'*10)

#[method 2 of doing the same thing]
for i in g:
    print(i)
```

```python
def infinite_generator(start):
    while True:
        start = start + 1
        yield start



g = infinite_generator(1)
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))




```

```python
l1 = [1,2,3,4,5]

for i in range(len(l1),0,-1):
    print(i)

def isOdd(i):
    if i%2 != 0:
        return True
    else:
        return False


def alternate_list(l):
    for i in range(len(l)):
        if isOdd(i):
            print(l[i])

print(alternate_list(l1))

```
